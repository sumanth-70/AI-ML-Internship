### Problem : 2

### Problem Statement :Print ManOfMach count of each player in Hyderabad stadium.

```python
import pandas as pd
df=pd.read_csv('DataFiles/Match.csv')

mom=[ ]
for i in range(len(df.values)):
    mom.append(df.values[i][14])

mom_hyd=[ ]
for i in range(len(df.values)):
    if(df.values[i][7]=='Hyderabad' and df.values[i][14]==mom[i]):
        mom_hyd.append(df.values[i][14])
unique_mom_hyd=[]
for i in range(len(df.values[i][14])):
    if mom_hyd not in unique_mom_hyd:
        unique_mom_hyd.append(mom_hyd[i])
unique_mom_hyd.sort()
no_of_matches=[ ]
for i in range(len(unique_mom_hyd)):
    no_of_matches.append(mom_hyd.count(unique_mom_hyd[i]))

df1=pd.DataFrame()
df1['Players']=unique_mom_hyd
df1['No.of MoMs in Hyderabad']=no_of_matches
df1
```

```python
import pandas as pd
def matcheswon(filename):
    df = pd.read_csv(filename)
    li=[]
    li1=[]
    a = df.groupby(['ManOfMach','City_Name']).count()
    for i in range(0,len(a)):
        if(str(a.index[i][1])== "Hyderabad"):
            li.append(a.iat[i,0])
            li1.append(a.index[i][0])
    d = {'Name':li1,'No of Match':li}
    df1 = pd.DataFrame(data = d)
    
    return df1
matcheswon('DataFiles/Match.csv')
```

### Problem : 3

### Problem statement :Print number of matches won and loss(as two columns). Consider as win when a team wins Toss and Match . print for each and every team.

```python
import pandas as pd
def matcheswon(filename):
    df = pd.read_csv(filename)
    li=[]
    li1=[]
    li2=[]
    a = df.groupby(['match_winner','Toss_Winner']).count()
    m = df.groupby(['Team1']).count()
    n = df.groupby(['Team2']).count()
    for i in range(0,len(a)):
        if(a.index[i][0]==a.index[i][1]):
            li.append(a.iat[i,0])
            li1.append(a.index[i][0])
    [li2.append(m.iat[i,0]+n.iat[i,0]-li[i]) for i in range(0,len(li1))]
    d = {'Name':li1,'win':li,'loss':li2}
    df1 = pd.DataFrame(data = d)
    return df1
matcheswon('DataFiles/Match.csv')
```
