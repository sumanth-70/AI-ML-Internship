```python
car_new = cars.drop(cars.index[[70,76]],axis = 0 ,inplace = False)
```

```python
print(m1_new.conf_int(0.05))
```

```python
mpg_pred = m1_new.predict(car_new)
```

```python
rsq_hp = smf.ols('HP ~ WT+VOL+SP',data = cars).fit()
rsq_hp.summary()
```

```python
rsq_hp = smf.ols('HP ~ WT+VOL+SP',data = cars).fit().rsquared
```

```python
vif_hp = 1/(1-rsq_hp)
```

```python
vif_hp
```

```python
rsq_sp = smf.ols('SP ~ WT + HP + VOL', data = cars).fit().rsquare
```

```python
vif_sp = 1/(1-rsq_sp)
vif_sp
```

```python
rsq_vol = smf.ols('VOL ~ WT + SP + HP',data =cars).fit().rsquared
```

```python
vif_vol =  1/(1-rsq_vol)
```

```python
vif_vol
```

```python
rsq_wt = smf.ols("WT ~ HP+SP+VOL", data = cars).fit().rsquared
```

```python
vif_wt = 1/(1-rsq_wt)
vif_wt
```
