### Problem : 1

### Problem statement:Draw a pie plot based on number of wins for each team.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
def Matchwin(filename):
    df = pd.read_csv(filename)
    x = df['match_winner'].value_counts()
    d = dict(x)
    li1=list(d.keys())           # Assigning the dictionary keys to list1
    li2=list(d.values())         # Assigning the dictioary values to list2
    labels = li1
    sizes = li2
    fig1, ax1 = plt.subplots()
    ax1.pie(sizes,labels=labels,autopct='%1.1f%%',shadow=True,startangle=90)
    ax1.axis('equal')
    plt.show()
    
Matchwin('DataFiles/Match.csv')
```
